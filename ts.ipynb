# Apps included on macOS

```python
load_url = 'https://support.apple.com/guide/mac-help/view-built-in-apps-mchl110b00b7/mac'
html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
def showApps(tbs, arr):
    for i in range(len(tbs)):
        # print(i)
        if i == 0:
            # print('Popular apps')
            arr.append('Popular apps')
        else:
            # print('Other apps')
            arr.append('Other apps')
        trs = tbs[i].find('tbody').find_all('tr')
        for j in range(len(trs)):
            # print(trs[j].find_all('td')[1].find('a').text)
            # print(trs[j].find_all('td')[1].find('a')['href']) # x-callback-url
            # print(trs[j].find_all('td')[1].find('a')['href']) # url
            # print('\t'+str(trs[j].find_all('td')[2].find('a').text.replace(' User Guide',''))+'\t'+str(trs[j].find_all('td')[2].find('a')['href']))
            arr.append('\t'+str(trs[j].find_all('td')[2].find('a').text.replace(' User Guide',''))+'\t'+str(trs[j].find_all('td')[2].find('a')['href']))
    return arr
```

```python
tbs = soup.find_all('table')
# results = '\n'.join(showApps(tbs, [])) # format
# print(results)
results = showApps(tbs, [])
```

```python
def showContents(lis, lvl, toc):
    for i in range(len(lis)):
        if lis[i].find_all('button', recursive=False):
            #print(lvl)
            #print(lis[i].findChildren('button', recursive=False)[0].text.replace('\n',''))
            toc.append(''.join(['\t' for i in range(lvl)])+lis[i].findChildren('button', recursive=False)[0].text.replace('\n',''))
            showContents(lis[i].find_all('li'), lvl+1, toc)
        else:
            #print(lvl)
            #print(lis[i].findChildren('a' , recursive=False)[0].text.replace('\n',''))
            #print(lis[i].findChildren('a' , recursive=False)[0]['href'])
            toc.append(''.join(['\t' for i in range(lvl)])+lis[i].findChildren('a' , recursive=False)[0].text.replace('\n','')+'\t'+lis[i].findChildren('a' , recursive=False)[0]['href'])
    return toc
```

```python
output = []
for i in range(len(results)):
    try:
        load_url = results[i].split('\t')[2]
        html = requests.get(load_url)
        soup = BeautifulSoup(html.content, "html.parser")
        lis = soup.find('ul', class_='hasIcons').find_all('li', recursive=False)
        output.append(results[i])
        output = output + showContents(lis, 2, [])
    except:
        output.append(results[i])
```

```python
# format
print('\n'.join(output))
```

```python
path_w = 'tstt.txt'
s = '\n'.join(output)

with open(path_w, mode='w') as f:
    f.write(s)
```

```python
# table
htm = ''
for i in range(len(output)):
    row = output[i].split('\t')
    htm += '<tr>'
    for j in range(len(row)):
        if j == len(row) - 1 and len(row) != 7:
            htm += '<td colspan="' + str(7-len(row)+1) + '">' + row[j] + '</td>'
        else:
            htm += '<td>' + row[j] + '</td>'
    htm += '</tr>\n'
print(htm)
```
